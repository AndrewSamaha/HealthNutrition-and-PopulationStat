# Section 4 Analyzing Iodized Salt 

```python

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 16, 'font.family': 'sans'}) 

health = pd.read_csv('data/data.csv')
```

```python
# What countires have iodized salt historical data
iodized_salt_record = health[health['Indicator Name'].str.contains("iodized salt")]
iodized_salt_record
```

```python
# The iodized salt consumption trend for each year
iodized_salt_yearly_mean = iodized_salt_record.mean(axis=1).loc['1993':'2013'] 
iodized_salt_yearly_mean
```

```python

x = np.arange(1993,2014)
y = iodized_salt_yearly_mean
fig, ax = plt.subplots(figsize=(10, 4))
ax.set_xlabel('Years')
ax.set_ylabel('Consumption of iodized salt (% of households)')
ax.plot(x, y, linewidth=2)
ax.set_title('Yearly Avg Consumption of iodized salt');
ax.set_xticks(x[::5]);
```
