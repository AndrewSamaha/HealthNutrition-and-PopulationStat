# Section 2. Analyzing Health Indicators

```python
heath_groupby = health.groupby(by= ['Indicator Name'])
```

```python
#Data Grouped by Different Types of Immunizations
immunization_groupby = health[health['Indicator Name'].str.contains("Immunization")].groupby(by= ['Indicator Name'])
immunization_groupby_mean = immunization_groupby.mean()
immunization_groupby_mean
```

```python
# Experiment with ggplot style
plt.style.use('ggplot') 
#set fonts and size
matplotlib.rcParams.update({'font.size': 25, 'font.family': 'sans'})
```

```python
x = immunization_groupby_mean.columns[:-1]
y = np.array(immunization_groupby_mean.iloc[0])[:-1]
y[np.isnan(y)] = 0
y_mean = immunization_groupby_mean.mean(axis=0)[:-1]
```

```python
fig, ax = plt.subplots(figsize=(40,15))

fig.suptitle('What is the Immunization trend in the world?', fontsize=40, fontweight='bold')

ax.plot(x, y, label='Immunization Mean', color='black', linewidth=5)
    
ax.set_title('Avg. World Immunization by Year')
ax.set_ylabel('Immunization Percent(%)')
ax.legend(loc="upper left")
ax.set_xticks(x[::10]);
```

```python
heath_groupby = health.groupby(by= ['Indicator Name']) 
immunization_groupby = health[health['Indicator Name'].str.contains("Immunization")].groupby(by= ['Indicator Name']) 
immunization_groupby.aggregate(np.mean)

countries = health["Country Name"].unique()
countries
health[health['Indicator Name'].str.contains("Immunization")]['Indicator Name'].unique()
```

```python
labels = immunization_groupby_mean.index
colors = ['blue','green','red','purple','orange','brown']
```

```python
fig, ax = plt.subplots(figsize=(40,15))

fig.suptitle('What does the Immunization breakdown trend looks like?', fontsize=40, fontweight='bold')
for i in range(immunization_groupby_mean.shape[0]):
    y = np.array(immunization_groupby_mean.iloc[i])[:-1]
    y[np.isnan(y)] = 0

    ax.plot(x, y, label=labels[i], color=colors[i], linewidth=5)
    
ax.set_title('Avg. World Immunization by Type')
ax.set_ylabel('Immunization Percent(%)')
ax.legend(loc="upper left")
ax.set_xticks(x[::10]);
```

```python
united_states_health = health[health['Country Code'] == 'USA']
united_states_immunization = united_states_health[united_states_health['Indicator Name'].str.contains("Immunization")]
united_states_immunization
```

```python
fig, ax = plt.subplots(figsize=(40,15))

fig.suptitle('What does the Immunization breakdown trend looks like in United States?', fontsize=40, fontweight='bold')
for i in range(united_states_immunization.shape[0]):
    y = np.array(united_states_immunization.iloc[i,4:])[:-1]

    ax.plot(x, y, label=labels[i], color=colors[i], linewidth=5)
    
ax.set_title('Avg. US Immunization by Type')
ax.set_ylabel('Immunization Percent(%)')
ax.legend(loc="upper left")
ax.set_xticks(x[::10]);
```

# Section 3. Condom Use and Rate of Children Born with HIV

```python
condom_groupby = health[health['Indicator Name'].str.contains("Condom")].groupby(by= ['Indicator Name']) 
condom_use = condom_groupby.aggregate(np.mean)
condom = condom_use.mean(axis=0)[:-1]
```

```python
condom_use.index
```

```python
HIV_groupby = health[health['Indicator Name'].str.contains("HIV")].groupby(by= ['Indicator Name']) 
children_w_HIV = HIV_groupby.mean().loc['Children (0-14) living with HIV'][:-1]
```

```python
fig, ax = plt.subplots(figsize=(40,15))
ax.scatter(children_w_HIV, condom, color="r", linewidths=10)
ax.set_title('Correlation Between Children Born with HIV and Condom Use')
ax.set_xlabel('# of Children Born w/ HIV')
ax.set_ylabel('% Children (0-14) living with HIV')
;
```
