```python
import numpy as np
import pandas as pd

# let's get matplot going
import matplotlib
import matplotlib.pyplot as plt  #standards should be followed
matplotlib.__version__
plt.style.use('ggplot') # Experiment with other styles to see for yourself which is best
matplotlib.rcParams.update({'font.size': 16, 'font.family': 'sans'})   #set fonts and size

health = pd.read_csv('data/data.csv')
years = health.columns[4:-1]
health_columns = np.array(health.count(axis=0))[4:-1]


print(health.iloc[100])
#print(health["Indicator Name"].unique())
#print(health["Indicator Code"].unique())
#print(health.info())


```

```python
fig, ax = plt.subplots(figsize=(10, 5))  


ax.plot(years, health_columns)
ax.set_title('Useful Data Points by Year')
ax.set_xlabel('Years')
ax.set_ylabel('Number of Useful Points')
ax.set_ylim(bottom=0)
ax.set_xticks(years[::10])

plt.tight_layout()
plt.show()
fig.savefig('figures/f1.usefuldatapointsbyyear.png')
```
