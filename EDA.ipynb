```python
# fig, ax = plt.subplots(figsize=(40,15))

# let's get matplot going
import matplotlib
import matplotlib.pyplot as plt  #standards should be followed
matplotlib.__version__
plt.style.use('ggplot') # Experiment with other styles to see for yourself which is best
matplotlib.rcParams.update({'font.size': 16, 'font.family': 'sans'})   #set fonts and size

health = pd.read_csv('data/data.csv')
years = health.columns[4:-1]
health_columns = np.array(health.count(axis=0))[4:-1]


print(health.iloc[100])
#print(health["Indicator Name"].unique())
#print(health["Indicator Code"].unique())
#print(health.info())


```

```python
fig, ax = plt.subplots(figsize=(10, 5))  


ax.plot(years, health_columns)
ax.set_title('Useful Data Points by Year')
ax.set_xlabel('Years')
ax.set_ylabel('Number of Useful Points')
ax.set_ylim(bottom=0)
ax.set_xticks(years[::10])

plt.tight_layout()
plt.show()
fig.savefig('figures/f1.usefuldatapointsbyyear.png')
```

```python
# Immunizations Averaged Across Time Per Country, by AS 
immunization_groupby = health[health['Indicator Name'].str.contains("Immunization")].groupby(by= ['Country Name']) 
immunization_groupby_mean = immunization_groupby.mean() 
per_country_rate = immunization_groupby_mean.transpose().mean(numeric_only=True).sort_values(na_position="first") 
per_country_rate_subset = per_country_rate[::10] 
per_country_rate_dropnan = per_country_rate.dropna()[::10] 
fig2, ax2 = plt.subplots(1,2,figsize=(10,10)) 
fig2.suptitle("Immunizations Averaged Across Time Per Country\n(1960-2015)", fontsize=18) 
fig2.text(0.5, .04, 'Immunization Rate', ha='center') 
ax2[0].barh(per_country_rate_subset.index, per_country_rate_subset) 
ax2[0].set_yticks(per_country_rate_subset.index) 
ax2[0].set_title('Including\nMissing Data', fontsize=12) 
ax2[1].set_yticklabels([]) 
ax2[1].set_yticks([]) 
ax3 = ax2[1].twinx() 
ax3.barh(per_country_rate_dropnan.index, per_country_rate_dropnan, align='center') 
ax2[1].set_title('Excluding\nMissing Data', fontsize=12) 
ax3.set_yticks(per_country_rate_dropnan.index) 
ax3.set_yticklabels(per_country_rate_dropnan.index) 
ax3.invert_xaxis() #plt.tight_layout() 
plt.show() 
fig2.savefig('figures/f2.immunizationspercountry.png', bbox_inches = "tight")
```

```python
# Experiment with ggplot style
plt.style.use('ggplot') 
#set fonts and size
matplotlib.rcParams.update({'font.size': 25, 'font.family': 'sans'})
```

```python
x = immunization_groupby_mean.columns[:-1]
y = np.array(immunization_groupby_mean.iloc[0])[:-1]
y[np.isnan(y)] = 0
y_mean = immunization_groupby_mean.mean(axis=0)[:-1]
```

```python
fig, ax = plt.subplots(figsize=(40,15))

fig.suptitle('What is the Immunization trend in the world?', fontsize=40, fontweight='bold')

ax.plot(x, y, label='Immunization Mean', color='black', linewidth=5)
    
ax.set_title('Avg. World Immunization by Year')
ax.set_ylabel('Immunization Percent(%)')
ax.legend(loc="upper left")
ax.set_xticks(x[::10])
fig.savefig('figures/f3.immunization.png');
```

```python
heath_groupby = health.groupby(by= ['Indicator Name']) 
immunization_groupby = health[health['Indicator Name'].str.contains("Immunization")].groupby(by= ['Indicator Name']) 
immunization_groupby.aggregate(np.mean)

countries = health["Country Name"].unique()
countries
health[health['Indicator Name'].str.contains("Immunization")]['Indicator Name'].unique()
```

```python
labels = immunization_groupby_mean.index
colors = ['blue','green','red','purple','orange','brown']
```

```python
fig, ax = plt.subplots(figsize=(40,15))

fig.suptitle('What does the Immunization breakdown trend looks like?', fontsize=40, fontweight='bold')
for i in range(immunization_groupby_mean.shape[0]):
    y = np.array(immunization_groupby_mean.iloc[i])[:-1]
    y[np.isnan(y)] = 0

    ax.plot(x, y, label=labels[i], color=colors[i], linewidth=5)
    
ax.set_title('Avg. World Immunization by Type')
ax.set_ylabel('Immunization Percent(%)')
ax.legend(loc="upper left")
ax.set_xticks(x[::10])
fig.savefig('figures/f4.immunization2.png');
```

```python
united_states_health = health[health['Country Code'] == 'USA']
united_states_immunization = united_states_health[united_states_health['Indicator Name'].str.contains("Immunization")]
united_states_immunization
```

```python
fig, ax = plt.subplots(figsize=(40,15))

fig.suptitle('What does the Immunization breakdown trend looks like in United States?', fontsize=40, fontweight='bold')
for i in range(united_states_immunization.shape[0]):
    y = np.array(united_states_immunization.iloc[i,4:])[:-1]

    ax.plot(x, y, label=labels[i], color=colors[i], linewidth=5)
    
ax.set_title('Avg. US Immunization by Type')
ax.set_ylabel('Immunization Percent(%)')
ax.legend(loc="upper left")
ax.set_xticks(x[::10])
fig.savefig('figures/f5.immunization3.png');
```

# Section 3. Analyzing Iodized Salt Consumption

```python
# What countires have iodized salt historical data
iodized_salt_record = health[health['Indicator Name'].str.contains("iodized salt")]
iodized_salt_record
```

```python
# The iodized salt consumption trend for each year
iodized_salt_yearly_mean = iodized_salt_record.mean(axis=0).loc['1993':'2013'] 
iodized_salt_yearly_mean
```

```python
# Plot the yearly trend
x = np.arange(1993,2014)
y = iodized_salt_yearly_mean
fig, ax = plt.subplots(figsize=(10, 4))
ax.set_xlabel('Years', fontsize=10)
ax.set_ylabel('Consumption of iodized salt (% of households)',fontsize=10)
ax.plot(x, y, linewidth=2)
ax.set_title('Yearly Avg Consumption of iodized salt');
ax.set_xticks(x[::5]);
```

```python
# The highest consumption every year
iodized_salt_yearly_max = iodized_salt_record.max(axis=0).loc['1993':'2013'] 
iodized_salt_yearly_max
```

```python
#Plot the highest cosumption

x = np.arange(1993,2014)
fig, ax = plt.subplots(figsize=(20, 8))
ax.bar(x, iodized_salt_yearly_max,color='g')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.set_ylabel('Iodized Salt Consumption')
ax.set_title('Yearly Highest Iodized Salt Consumption 1993-2013 ') 
length = len(iodized_salt_yearly_max)
```
